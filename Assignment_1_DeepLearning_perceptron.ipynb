```python
x=df.drop(labels='1',axis=1)
y=df['1']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=32)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler_fit=scaler.fit_transform(x)
```

### Perceptron

```python
from sklearn.linear_model import Perceptron
per=Perceptron()
per.fit(x_train,y_train)
per_pred=per.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,f1_score,recall_score,confusion_matrix
```

```python
print('accuracy_score:',accuracy_score(y_test,per_pred))
print('precision_score:',precision_score(y_test,per_pred))
print('f1_score:',f1_score(y_test,per_pred))
print('recall_score:',recall_score(y_test,per_pred))
print('confusion_matrix\n:',confusion_matrix(y_test,per_pred))
```

```python
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
```

```python
# define per evaluation method
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate
scores = cross_val_score(per, x, y, scoring='accuracy', cv=cv, n_jobs=-1)
# summarize result
print('Mean Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```

### Fine Tuning

```python
grid=dict()
grid['eta0']=[0.0001,0.001,0.01,0.1,1.0]
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import GridSearchCV,RepeatedStratifiedKFold
from sklearn.linear_model import Perceptron

x,y=make_classification(n_samples=1000,n_features=10,n_informative=10,n_redundant=0,random_state=1)
per=Perceptron()
cv=RepeatedStratifiedKFold(n_splits=10,n_repeats=3,random_state=1)
grid=dict()
grid['eta0']=[0.0001,0.001,0.01,0.1,1.0]
search=GridSearchCV(per, grid, scoring='accuracy', cv=cv, n_jobs=-1)
results=search.fit(x,y)
print('Mean Accuracy: %.3f' % results.best_score_)
print('Config: %s' % results.best_params_)
means = results.cv_results_['mean_test_score']
params = results.cv_results_['params']
for mean, param in zip(means, params):
    print(">%.3f with: %r" % (mean, param))
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import GridSearchCV,RepeatedStratifiedKFold
from sklearn.linear_model import Perceptron

x,y=make_classification(n_samples=1000,n_features=10,n_informative=10,n_redundant=0,random_state=1)
per=Perceptron()
cv=RepeatedStratifiedKFold(n_splits=10,n_repeats=3,random_state=1)
grid = dict()
grid['max_iter'] = [1, 10, 100, 1000, 10000]
search=GridSearchCV(per, grid, scoring='accuracy', cv=cv, n_jobs=-1)
results=search.fit(x,y)
print('Mean Accuracy: %.3f' % results.best_score_)
print('Config: %s' % results.best_params_)
means = results.cv_results_['mean_test_score']
params = results.cv_results_['params']
for mean, param in zip(means, params):
    print(">%.3f with: %r" % (mean, param))
```
